***Etape 3 : Recuperation des informations des produits avec "BeautifulSoup"***

```python
from bs4 import BeautifulSoup

# Charger le fichier HTML
with open('jumia_electronique.html', 'r', encoding='utf-8') as file:
    contenu_html = file.read()

# Analyser le contenu HTML
soup = BeautifulSoup(contenu_html, 'html.parser')

# Initialiser des listes pour stocker les données extraites
noms_produits = []
images_produits = []
prix_produits = []

# Trouver tous les conteneurs de produits
produits = soup.find_all('div', class_='itm col')

# Boucler à travers chaque produit et extraire les informations
for produit in produits:
    # Extraire le nom du produit
    balise_nom = produit.find('div', class_='name')
    if balise_nom:
        noms_produits.append(balise_nom.text.strip())

    # Extraire l'URL de l'image du produit
    balise_image = produit.find('img', class_='img')
    if balise_image:
        images_produits.append(balise_image.get('data-src') or balise_image.get('src'))

    # Extraire le prix du produit
    balise_prix = produit.find('div', class_='prc')
    if balise_prix:
        prix_produits.append(balise_prix.text.strip())

# Vérifier le nombre de produits extraits
print(f"Nombre de produits extraits : {len(noms_produits)}")

# Afficher les données extraites
for nom, image, prix in zip(noms_produits, images_produits, prix_produits):
    print(f'Nom : {nom}\nImage : {image}\nPrix : {prix}\n')
```

### ***<p style="font-style:italic; text-decoration:underline;">Gesion des erreurs avec BeautifulSoup et requests***</p>

***Pour mettre en place des mécanismes de gestion des erreurs dans notre projet de web scraping avec "BeautifulSoup" et "requests", nous pouvons ajouter des vérifications pour gérer les temps de réponse, les pages manquantes et autres erreurs potentielles. Voici comment nous pouvons intégrer ces mécanismes***

```python
import requests
from bs4 import BeautifulSoup
import time

def fetch_html(url, retries=3, delay=5):
    """
    Récupère le contenu HTML d'une URL avec gestion des erreurs.
    
    Parameters:
    url (str): URL de la page à scraper.
    retries (int): Nombre de tentatives avant d'abandonner.
    delay (int): Temps d'attente entre les tentatives en secondes.
    
    Returns:
    str: Contenu HTML de la page.
    """
    for attempt in range(retries):
        try:
            response = requests.get(url, timeout=10)
            response.raise_for_status()  # Vérifie si la requête a réussi
            return response.text
        except requests.exceptions.RequestException as e:
            print(f"Erreur lors de la récupération de l'URL (tentative {attempt + 1}/{retries}) : {e}")
            time.sleep(delay)  # Attendre avant de réessayer
    return ""

def extract_product_info(soup):
    """
    Extrait les informations des produits depuis le contenu HTML.
    
    Parameters:
    soup (BeautifulSoup): Objet BeautifulSoup contenant le contenu HTML analysé.
    
    Returns:
    tuple: Trois listes contenant les noms des produits, les URL des images et les prix.
    """
    noms_produits = []
    images_produits = []
    prix_produits = []

    # Trouver tous les conteneurs de produits
    produits = soup.find_all('div', class_='itm col')

    # Boucler à travers chaque produit et extraire les informations
    for produit in produits:
        try:
            # Extraire le nom du produit
            balise_nom = produit.find('div', class_='name')
            if balise_nom:
                noms_produits.append(balise_nom.text.strip())
            else:
                noms_produits.append('N/A')
                print("Nom non trouvé pour un produit")

            # Extraire l'URL de l'image du produit
            balise_image = produit.find('img', class_='img')
            if balise_image:
                images_produits.append(balise_image.get('data-src') or balise_image.get('src'))
            else:
                images_produits.append('N/A')
                print("Image non trouvée pour un produit")

            # Extraire le prix du produit
            balise_prix = produit.find('div', class_='prc')
            if balise_prix:
                prix_produits.append(balise_prix.text.strip())
            else:
                prix_produits.append('N/A')
                print("Prix non trouvé pour un produit")
        except Exception as e:
            print(f"Erreur lors de l'extraction d'un produit : {e}")
            noms_produits.append('N/A')
            images_produits.append('N/A')
            prix_produits.append('N/A')

    return noms_produits, images_produits, prix_produits

# URL de la page à scraper
url = 'https://www.jumia.sn/electronique/'

# Récupérer le contenu HTML de la page
html_content = fetch_html(url)

if html_content:
    # Analyser le contenu HTML
    soup = BeautifulSoup(html_content, 'html.parser')
    
    # Extraire les informations des produits
    noms_produits, images_produits, prix_produits = extract_product_info(soup)
    
    # Vérifier le nombre de produits extraits
    print(f"Nombre de produits extraits : {len(noms_produits)}")

    # Afficher les données extraites
    for nom, image, prix in zip(noms_produits, images_produits, prix_produits):
        print(f'Nom : {nom}\nImage : {image}\nPrix : {prix}\n')
else:
    print("Échec de la récupération du contenu HTML après plusieurs tentatives.")
```

### ***<p style="font-style:italic; text-decoration:underline;">Stockage des donnees dans un fichier CSV</p>***

***Pour stocker les données extraites dans un fichier CSV, nous pouvons utiliser le module "csv" de Python. Voici comment nous pouvons ecrire le code pour enregistrer les informations des produits dans un fichier CSV***

```python
import csv
from bs4 import BeautifulSoup

# Charger le fichier HTML
with open('jumia_electronique.html', 'r', encoding='utf-8') as file:
    contenu_html = file.read()

# Analyser le contenu HTML
soup = BeautifulSoup(contenu_html, 'html.parser')

# Initialiser des listes pour stocker les données extraites
noms_produits = []
images_produits = []
prix_produits = []

# Trouver tous les conteneurs de produits
produits = soup.find_all('div', class_='itm col')

# Boucler à travers chaque produit et extraire les informations
for produit in produits:
    # Extraire le nom du produit
    balise_nom = produit.find('div', class_='name')
    if balise_nom:
        noms_produits.append(balise_nom.text.strip())
    else:
        noms_produits.append('N/A')

    # Extraire l'URL de l'image du produit
    balise_image = produit.find('img', class_='img')
    if balise_image:
        images_produits.append(balise_image.get('data-src') or balise_image.get('src'))
    else:
        images_produits.append('N/A')

    # Extraire le prix du produit
    balise_prix = produit.find('div', class_='prc')
    if balise_prix:
        prix_produits.append(balise_prix.text.strip())
    else:
        prix_produits.append('N/A')

# Vérifier le nombre de produits extraits
print(f"Nombre de produits extraits : {len(noms_produits)}")

# Enregistrer les données extraites dans un fichier CSV
with open('produits_jumia.csv', 'w', newline='', encoding='utf-8') as csvfile:
    fieldnames = ['Nom', 'Image', 'Prix']
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)

    writer.writeheader()
    for nom, image, prix in zip(noms_produits, images_produits, prix_produits):
        writer.writerow({'Nom': nom, 'Image': image, 'Prix': prix})

print("Les données ont été enregistrées dans le fichier produits_jumia.csv")

# Lire et afficher les données du fichier CSV
# print("\nProduits stockés dans le fichier CSV :")
# with open('produits_jumia.csv', 'r', encoding='utf-8') as csvfile:
#     reader = csv.DictReader(csvfile)
#     for row in reader:
#         print(f"Nom : {row['Nom']}\nImage : {row['Image']}\nPrix : {row['Prix']}\n")
```

***1. Préparation des Données***

- ***Nettoyer les données en supprimant les valeurs manquantes et en convertissant les prix en un format numérique.***

```python
import pandas as pd

# Charger les données depuis le fichier CSV
data = pd.read_csv('produits_jumia.csv')

# Nettoyer les données
# Remplacer 'N/A' par NaN
data.replace('N/A', pd.NA, inplace=True)

# Supprimer les lignes avec des valeurs manquantes
data.dropna(inplace=True)

# Fonction pour nettoyer et convertir les prix
def clean_price(price):
    # Supprimer 'FCFA', les virgules et les espaces
    price = price.replace('FCFA', '').replace(',', '').replace(' ', '')
    # Vérifier si le prix est une plage
    if '-' in price:
        low, high = map(float, price.split('-'))
        price = (low + high) / 2
    else:
        price = float(price)
    return price

# Appliquer la fonction de nettoyage à la colonne 'Prix'
data['Prix'] = data['Prix'].apply(clean_price)

# Afficher les premières lignes des données nettoyées
print(data.head())
```

***2. Analyse des Données***

- ***Analyser les données en calculant des statistiques descriptives, comme le prix moyen, le prix minimum et le prix maximum.***
- ***Identifier les produits les plus chers et les moins chers.***

```python
# Calculer des statistiques descriptives
prix_mean = data['Prix'].mean()
prix_min = data['Prix'].min()
prix_max = data['Prix'].max()
print("### Calcul des statistiques descriptives ###")
print(f"Prix moyen : {prix_mean} FCFA")
print(f"Prix minimum : {prix_min} FCFA")
print(f"Prix maximum : {prix_max} FCFA")

# Trouver les produits les plus chers et les moins chers
produit_cher = data.loc[data['Prix'].idxmax()]
produit_pas_cher = data.loc[data['Prix'].idxmin()]

print("\nProduit le plus cher :")
print(produit_cher)
print("\nProduit le moins cher :")
print(produit_pas_cher)
```

***3. Visualisation***

- ***Créer des visualisations pour illustrer les résultats de l'analyse, comme des histogrammes des prix des produits, des diagrammes en barres des catégories de produits et des nuages de points des prix des produits***

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Charger les données depuis le fichier CSV
data = pd.read_csv('produits_jumia.csv')

# Nettoyer les données
data.replace('N/A', pd.NA, inplace=True)
data.dropna(inplace=True)

def clean_price(price):
    price = price.replace('FCFA', '').replace(',', '').replace(' ', '')
    if '-' in price:
        low, high = map(float, price.split('-'))
        price = (low + high) / 2
    else:
        price = float(price)
    return price

data['Prix'] = data['Prix'].apply(clean_price)

# Histogramme des prix des produits
plt.figure(figsize=(10, 6))
sns.histplot(data['Prix'], kde=True)
plt.title('Distribution des Prix des Produits')
plt.xlabel('Prix (FCFA)')
plt.ylabel('Fréquence')
plt.show()

# Diagramme en barres des 10 produits les plus chers
top_10_produits = data.nlargest(10, 'Prix')

plt.figure(figsize=(10, 6))
sns.barplot(x='Prix', y='Nom', data=top_10_produits, palette='viridis')
plt.title('Top 10 des Produits les Plus Chers')
plt.xlabel('Prix (FCFA)')
plt.ylabel('Nom du Produit')
plt.show()

# Nuage de points des prix des produits
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Nom', y='Prix', data=data)
plt.xticks(rotation=90)
plt.title('Nuage de Points des Prix des Produits')
plt.xlabel('Nom du Produit')
plt.ylabel('Prix (FCFA)')
plt.show()
```
